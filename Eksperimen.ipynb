```python
import numpy as np
import pandas as pd

df_train = pd.read_csv('train.csv', sep=',')
df_test = pd.read_csv('test.csv', sep=',')

df_train
```

```python
# EDA : check null and ? value

print("EDA train data")
print("Null value in train data:")
print(df_train.isnull().sum())
print("\n")
for column in df_train.columns:
    print(df_train[column].value_counts())
print("===============================\n")
print("EDA test data")
print("Null value in test data:")
print(df_test.isnull().sum())
print("\n")
for column in df_test.columns:
    print(df_test[column].value_counts())

# So many '?' values at those columns, drop them
df_train.drop(['slope', 'ca', 'thal'], axis = 1, inplace=True)
df_test.drop(['slope', 'ca', 'thal'], axis = 1, inplace=True)
```

```python
# Handle weird disgusting formatting for num column in df_test
print(df_train.columns)
print(df_test.columns)
df_test['num'] = df_test['num       ']
df_test.drop(['num       '], axis = 1, inplace=True)
```

```python
df_test
```

```python
# # Replace '?' for remaining column with each corresponding mean column val
for column in df_train.columns:
    df_train = df_train[df_train[column] != '?']

for column in df_test.columns:
    df_test = df_test[df_test[column] != '?']
```

```python
X_train = df_train.drop(['num'], axis = 1)
y_train = df_train['num']
X_test = df_test.drop(['num'], axis = 1)
y_test = df_test['num']

print(X_train)
print(X_test)
```

```python
X_combine = pd.concat([X_train, X_test])
y_combine = pd.concat([y_train, y_test])

print(X_combine)
print(y_combine)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 2, random_state = 1).fit(X_train)
kmeans2 = KMeans(n_clusters = 2, random_state = 1).fit(X_combine)

kmeans_pred = kmeans.labels_
kmeans2_pred = kmeans2.labels_
```

```python
def normalize(data):
    array = []
    for datum in data:
        array.append(datum)
    return array

pred = normalize(kmeans_pred)
target = normalize(y_train)
pred2 = normalize(kmeans2_pred)
target2 = normalize(y_combine)
```

```python
def purity(pred, target):
    purity_matrix = [[0,0],[0,0]]
    for i in range (len(target)):
        purity_matrix[pred[i]][target[i]] = purity_matrix[pred[i]][target[i]] + 1
    first_guest = (purity_matrix[0][0] + purity_matrix[1][1])
    second_guest = (purity_matrix[0][1] + purity_matrix[1][0])
    print(purity_matrix)
    return f'Purity = {( max(first_guest , second_guest)) / len(target)}'

print("Purity of K Means only train data : ")
print(purity(pred, target))

print("Purity of K Means train+test data : ")
print(purity(pred2, target2))
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=100, min_samples=2).fit(X_train)
dbscan2 = DBSCAN(eps=100, min_samples=2).fit(X_combine)

pred1 = normalize(dbscan.labels_)
target1 = normalize(y_train)
pred2 = normalize(dbscan2.labels_)
target2 = normalize(y_combine)

print("Purity of DBSCAN only train data : ")
print(purity(pred1, target1))

print("Purity of DBSCAN train+test data : ")
print(purity(pred2, target2))
```
